```python
#configurações iniciais
from src.configs.paths import PROCESSED_DATA_DIR
from src.io.io_utils import load_parquet
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
```

```python
path_dataset = PROCESSED_DATA_DIR / "cicids2017/friday.parquet"
df = load_parquet(path_dataset)
```

```python
#estatísticas descritivas (média, mediana, desvio padrão, min, max)
# describe oferece (média, descio padrão - std, min, max, quartis )
"""
conceitos
std - desvio padrão: mede o quanto os valores variam da média, raiz da variância, mede a média das distâncias dos valores até a média. indica se os dados estão concentrados ou espalhados

std alto - pode significar que é informativa, mas tomar cuidado com outliers # histograma e análise por classe para verificar
std baixo - pouca variação, quase constante, candidato a remoção.

------ baixa variação significa que não possui influência na variável independente (verificar com michael)

levar em consideração:

* media aproximada de 0
    - ignorar cv
    - valores constantes ou esparsos

* std baixo
    - pouca variação
    - provável remoção

* std alto em relação a média
    - alta dispersão
    - buscar por outliers, caudaa longa (estudar conceito)



    tomar cuidado com média próxima a 0, já que a divisão por um número pequeno vai dar um resultado maior
    ignorar cv quando mean < 0.01



"""
pd.options.display.float_format = '{:.2f}'.format

desc = df.describe().T
desc = desc.drop(columns=['count'])
#fórmula -> cv = std / mean
desc['cv'] = desc['std'] / desc['mean']
#exibir das mais instáveis para mais constantes
desc = desc.sort_values(by='cv', ascending=False)
display(desc)


pd.reset_option('display.float_format')
```
