```python
from src.configs.paths import PROCESSED_DATA_DIR
from src.io.io_utils import load_parquet
import pandas as pd
import numpy as np
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
path_dataset = PROCESSED_DATA_DIR / "cicids2017/friday.parquet"
df = load_parquet(path_dataset)
```

```python

# Seleção de feature
feature_selector = widgets.Dropdown(
    options=df.select_dtypes(include='number').columns.tolist(),
    description='Feature:',
)

# Tipo de gráfico
plot_type = widgets.Dropdown(
    options=['Histograma', 'Boxplot'],
    description='Gráfico:',
)

scale_type = widgets.Dropdown(
    options=['Normal', 'Log (dados)'],
    description='Escala:',
)

# Botão
button = widgets.Button(description="Gerar")

# Output
output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output(wait=True)
        feature = feature_selector.value

        df_plot = df.copy()

        # 🔥 transformação correta
        if scale_type.value == 'Log (dados)':
            df_plot[feature] = np.log1p(df_plot[feature])
            title_suffix = " (log1p)"
        else:
            title_suffix = ""

        if plot_type.value == 'Histograma':
            fig = px.histogram(
                df_plot,
                x=feature,
                nbins=50,
                title=f'Histograma{title_suffix} - {feature}'
            )

        elif plot_type.value == 'Boxplot':
            fig = px.box(
                df_plot,
                x=feature,
                title=f'Boxplot{title_suffix} - {feature}'
            )

        fig.show()

button.on_click(on_button_click)

display(feature_selector, plot_type, scale_type, button, output)

#src2dst_bytes
```
